````
## Creating a function to download img from a url specified by the USER
import urllib.request as urlrequest

def dl_img(url, file_path, file_name):
  _path = file_path + file_name + '.jpg'
  urlrequest.urlretrieve(url, _path)
  
url = input("URL: ")
saveas = input("File Name: ")

dl_img(url, '/content/', saveas)
````

> img = cv.imread(path_to_img, filter)

filter : 
   
  * cv.IMREAD_COLOR - Loads color image without alpha channel, default (1)
   
  * cv.IMREAD_GRAYSCALE - Loads color image as grayscale (0)
   
  * cv.IMREAD_UNCHANGED - Loads color image with alpha channel (-1)

> What is ALPHA?
  * degree of opaqueness
  * how much less transparent it is 

* Why GRAYSCALE preferred??
> (r, g, b) but in cv2:  (b, g, r)

  > (b, g, r, alpha): 4 channels -- processing more =>slower
