```
## Creating a function to download img from a url specified by the USER
import urllib.request as urlrequest

def dl_img(url, file_path, file_name):
  _path = file_path + file_name + '.jpg'
  urlrequest.urlretrieve(url, _path)
  
url = input("URL: ")
saveas = input("File Name: ")

dl_img(url, '/content/', saveas)

# https://cdn.shopify.com/s/files/1/1893/0477/products/5PCS_Framed_Colorful_Lion_Canvas_Prints_grande.png?v=1504331897
```

```
# region of the img
roi = img1[100:150, 100:150]

print(type(roi))

# convert all the pixels in that region to white(255,255,255) 
# incorrect method...doesn't work for the complete region values  
# roi = [255,255,255]
img1[100:150, 100:150] = [255,255,255]

# display the pixel values of the region (array format)  
print(roi)

# display the img.
cv_imshow(cv.resize(img1, None, None, 0.5, 0.5))

```

```
## COPY + PASTE a region

# SELECT/COPY a region 
nose = img1[250:350, 250:350]

# PASTE it
img1[0:100, 0:100] = nose

# display the img.
cv_imshow(cv.resize(img1, None, None, 0.5, 0.5))
```
